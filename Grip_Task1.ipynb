# Task1: Predict the percentage of an student based on the no. of study hours.


```
# Import Required Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
# Uploading the File to Google Colab from your local Computer
from google.colab import files
upload = files.upload()
```

```
# Read the Dataset
dataset=pd.read_csv("dataset.csv")
```

```
# Printing our Imported Dataset
print("Our Dataset Imported is as: \n",dataset)
```

```
#Present the shape of our Dataset i.e. 25 Rows and 2 Coloumns 
dataset.shape
```

```
dataset.info()
```

```
dataset.describe()
```

```
# Grouping our Dataset with respect to their respective mean values
dataset.groupby(['hours'])['scores'].mean()
```

## **Plot the Dataset**

```
plt.scatter(dataset['hours'],dataset['scores'],color='Red', marker='o')
plt.title("Scatter plot: Hours vs Scores")
plt.ylabel("Scores")
plt.xlabel("Hours")
plt.show()
```

```
correlation=dataset.corr()
correlation
```

```
sns.regplot(data= dataset,x= 'hours',y='scores')
plt.title("Plotting Data points over the line")
```

```
X= dataset.iloc[:, :-1].values
Y= dataset.iloc[:,-1].values
```

## **Splitting Our Dataset into Testing and Training Sets**

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=1/3, random_state= 0)
```

## ***Import Regression model from Sklearn Lobrary***

```
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

## **Testing our Model on our testing set**

```
y_pred= regressor.predict(X_test)
y_pred
```

```
pred_data = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
pred_data
```

## **Visualizing our Training Data Result**

```
plt.scatter(X_train, y_train, color='Red')
plt.plot(X_train, regressor.predict(X_train), color='Green')
plt.title("Plotting for Training Data")
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

## **Visualizing our Models Test Set Result**

```
plt.scatter(X_test, y_test, color='Red')
plt.plot(X_train, regressor.predict(X_train), color='Green')
plt.title("Plotting for Training Data")
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

# **Predicting Output for New Value on our Trained Model**

```
new_value = float(input("Enter value to predict: "))
```

```
new_value = np.array(new_value)
new = new_value.reshape(1,-1)
task_pred = regressor.predict(new)
print("Predicted Score for the", new_value , "is: ", task_pred)
```
